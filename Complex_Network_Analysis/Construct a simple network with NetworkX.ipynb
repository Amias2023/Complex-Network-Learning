# 添加 边和节点

* 可以一个一个添加，也可以从一个列表或者另一个 graph 中添加。

* 同样的可以一个一个的移除，或者使用一个列表

**规则：**

* 移除一个边不移除两边的节点

* 移除一个节点移除所有相关的边

* 添加一个重复的节点或者边默认忽略（除非是 multigraph）

* 添加一个边，如果边的两个节点之前不存在，也同时添加节点

## 创建一个事物和营养物质的网络 -- toy

```python
G = nx.Graph([("A", "eggs")],)
```

### 创建边和节点

```python
G.add_node("spinach") # 添加一个节点
G.add_node("mistake") # 添加一个错误的节点
G.add_nodes_from(["folates","asparagus","liver"]) # 添加一列表的节点
G.add_edge("spinach", "folates") # 添加一个两个节点都存在的边
G.add_edge("spinach","mistake_edge") # 添加一个错误的连边
G.add_edges_from([("folates","liver"),("folates", "asparagus")]) # 添加一列表的连边

# 删除错误的节点和连边

G.remove_node("mistake")
G.remove_edge("spinach","mistake_edge")

G.remove_node("mistake_edge") # 删除边并不能删除这个节点


# G.clear()  # 删除所有的节点和连边
```

### 重新 label 每个节点

* 例如将节点的名字全部大写

```python
mapping = {node : node.title() for node in G if isinstance(node, str)}
nx.relabel_nodes(G,mapping, copy = False)
print(G.nodes())
```

* 使用一个字典表示替换：h the old labels as keys and new labels as values.

## 添加属性数据

* 边的属性信息： 权重，强度， 吞吐量

* 节点的属性信息： 描述节点的信息

----

* 属性信息储存在字典中， 通过字典进行添加

### 在添加节点和边的时候定义属性

```python
G.add_node("Honey", edible = True)
G.add_nodes_from([("Steel", {"edible": False})]) # 字典的 key 是属性， value 是相应的属性值

G.add_edge("Honey", "Steel", weight = 0) # 连边的权重是 0 
G.add_edges_from(["Honey","Zn"], related = False) # 所有的连边都添加相同的属性

# 许多复杂网络分析的库 都是统一使用 weight 来表示连边的强度

# 直接添加带权重的边, 有权图

G.add_weighted_edges_from([("Honey", "Zn", 0.01), ("Honey", "Suger", 0.99)])
```

### 通过字典的 label 修改节点的属性

```python
G.node["Zn"]["nutrient"] = True #  更改节点 Zn 的 nutrient 属性的值
G.node["Zn"]["Beef"]["Weight"] = 0.95 # 更改 Zin 和 Beef 之间 edge 的 Weight 属性的值

# 删除 相应的属性的值
del G.node["Zn"]["nutrient"] 
del G.node["Zn"]["Beef"]["Weight"]
```

* 将我们创建的网络中的 节点进行标记 ，添加一个属性： 是否是 nutrient


```python
nutrients = set(("B12", "Zn", "D", "B6", "A", "Se", "Cu", "Folates", "Ca",
                 "Mn", "Thiamin", "Riboflavin", "C", "E", "Niacin"))
nutrient_dict = {node: (node in nutrients) for node in G}
```

```python
nx.set_node_attributes(G, nutrient_dict, name= "nutrient")

# nx.set_node_attributes(G, values, name=None)
```

## 可视化网络

* 可视化的过程主要包括两个方面： 布局和渲染

* NetworkX 提供一系列布局的算法 （选择节点的地理位置）

    * random
    
    * Circular ： 节点在一个圆上
    
    * Shell ：  同心圆
    
    * Specttal ： 基于特征向量中心值
    
    * force-directed : 好像通过弹簧连接的物理球互相排斥
    
* rendering 那一步， NetworkX 绘制节点，标签，边在相应的位置
 
* All-in-one function 既可以布局又可以渲染。

* 之后进行保存

###  为节点选定颜色

```python
colors = ["yellow" if n[1]["nutrient"] else "pink" for n in G.nodes(data = True)]

import  dzcnapy_plotlib as dzcnapy
dzcnapy.medium_attrs["node_color"] = colors
```

### 四种布局

```python

_, plot = plt.subplots(2, 2)
subplots = plot.reshape(1, 4)[0]
layouts = (nx.random_layout, nx.circular_layout, nx.spring_layout,
           nx.spectral_layout)
titles = ("Random", "Circular", "Force-Directed", "Spectral")

for plot, layout, title in zip(subplots, layouts, titles):
    pos = layout(G)
    
    # 调用 nx.draw_networkx()
    
    nx.draw_networkx(G,pos=pos, ax = plot, with_labels= False, **dzcnapy.medium_attrs)
    plot.set_title(title)
    dzcnapy.set_extent(pos,plot)
dzcnapy.plot("nutrients")
```

```python
nx.draw_networkx()
```

* 看起来， spring layout ， 第三个，有最漂亮的输出

### 使用 Graphviz 进行可视化

```python
from networkx.drawing.nx_agraph import graphviz_layout
_,plot = plt.subplots()
pos = graphviz_layout(G)
nx.draw_networkx(G, pos, **dzcnapy.attrs)
dzcnapy.set_extent(pos, plot)

dzcnapy.plot("nutrient_graphviz")
```
